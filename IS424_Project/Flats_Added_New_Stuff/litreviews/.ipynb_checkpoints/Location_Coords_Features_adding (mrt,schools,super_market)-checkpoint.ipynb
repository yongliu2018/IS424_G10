# Adding nearest MRT

```python
df = pd.read_csv('resale-flat-prices-based-on-registration-date-from-jan-2017-onwards.csv')
```

```python
mrt = pd.read_excel('mrt.xlsx')
mrt.head()
```

```python
df = pd.read_csv('resale-flat-prices-based-on-registration-date-from-jan-2017-onwards.csv')
df['month']=pd.to_datetime(df['month'])
df['Year'] = df['month'].apply(lambda x: x.year)
df['Month'] = df['month'].apply(lambda y: y.month)
df = df.drop(columns = ['month','Month'])
df = df[df['Year'] == 2019]
df.head()
```

```python
df['street_name'] = df['street_name'].replace(["ST. GEORGE'S RD"],"ST GEORGE'S RD")
df['street_name'] = df['street_name'].replace(["ST. GEORGE'S LANE"],"ST GEORGE'S LANE")
```

```python
df['address'] = df['block'] + " " + df['street_name']
df = df.drop(columns = ['block','street_name'])
```

```python
df_unique_address = df['address'].drop_duplicates()
df_unique_address.reset_index(drop=True, inplace=True)
len(df_unique_address)
```

```python
# This cell ran for >30 min to retrieve 8k+ records

dict1 = {}

for row in range(len(df_unique_address)):
    query_address = df_unique_address[row]
    query_string = 'https://developers.onemap.sg/commonapi/search?searchVal=' + str(query_address) + '&returnGeom=Y&getAddrDetails=Y'
    resp = requests.get(query_string)

    # Convert JSON into Python Object
    data_geo_location = json.loads(resp.content)
    if data_geo_location['found'] != 0:
        dict1[query_address] = [data_geo_location['results'][0]['LATITUDE'],
                                data_geo_location['results'][0]['LONGITUDE'],
                                data_geo_location['results'][0]['POSTAL'],
                                data_geo_location['results'][0]['ADDRESS']
                               ]
```

```python
df["Latitude"] = np.nan
df["Longtitude"] = np.nan
df["Postal"] = np.nan
df["full_address"] = np.nan

def assign(address):
    if address in dict1:
        x = dict1[address][0]
        return x
df['Latitude'] = df['address'].apply(assign)

def assign(address):
    if address in dict1:
        x = dict1[address][1]
        return x
df['Longtitude'] = df['address'].apply(assign)

def assign(address):
    if address in dict1:
        x = dict1[address][2]
        return x
df['Postal'] = df['address'].apply(assign)

def assign(address):
    if address in dict1:
        x = dict1[address][3]
        return x
df['full_address'] = df['address'].apply(assign)
```

```python
# get the nearest mrt
from geopy.distance import geodesic

def find_nearest(add_lat_lon_start,add_lat_lon_end):
    
    results = {}
    # first column must be address
    for index,flat in enumerate(add_lat_lon_start.iloc[:,13]):
        
        # 2nd column must be latitude, 3rd column must be longitude
        flat_loc = (add_lat_lon_start.iloc[index,10],add_lat_lon_start.iloc[index,11])
        nearest_stn = ['',10]
        for ind, stn in enumerate(add_lat_lon_end.iloc[:,1]):
            stn_loc = (add_lat_lon_end.iloc[ind,5],add_lat_lon_end.iloc[ind,6])
            distance = geodesic(flat_loc,stn_loc).km

            if distance < nearest_stn[1]:
                nearest_stn[0] = stn
                nearest_stn[1] = distance

        results[flat] = nearest_stn
        
    return results

a = find_nearest(df,mrt)
a
# a['330 ANG MO KIO AVENUE 1 TECK GHEE VIEW SINGAPORE 560330']
```

```python
def assign(dist):
    if dist in a:
        x = a[dist][0]
        return x
df['nearest_mrt'] = df['full_address'].apply(assign)

def assign(dist):
    if dist in a:
        x= a[dist][1]
        return x
df['nearest_mrt_distance(km)'] = df['full_address'].apply(assign)
```

```python
df.sample(10)
```

```python
df.to_csv('2019_hdb_dataset_with_mrt.csv')
```

# Adding Nearest  school

```python
new = pd.read_csv('2019_hdb_dataset_with_mrt.csv')
schools = pd.read_excel('schools.xlsx')

schools = pd.read_excel('schools.xlsx', converters={'postal_code': lambda x: str(x)})
```

```python
schools['mainlevel_code'].unique()
```

```python
primary = schools[schools['mainlevel_code'] == 'PRIMARY']
secondary = schools[schools['mainlevel_code'] == 'SECONDARY']
mixed = schools[schools['mainlevel_code'] == 'MIXED LEVELS']
JC_MI = schools[(schools['mainlevel_code'] == 'JUNIOR COLLEGE') | (schools['mainlevel_code'] == 'CENTRALISED INSTITUTE')]
```

```python
table = [primary,secondary,mixed,JC_MI]
for i in table:
    geo  = {}
    missing = []
    for row in i['postal_code']:
        query_address = row
        query_string = 'https://developers.onemap.sg/commonapi/search?searchVal=' + str(query_address) + '&returnGeom=Y&getAddrDetails=Y'
        resp = requests.get(query_string)

        # Convert JSON into Python Object
        data_geo_location = json.loads(resp.content)
        if data_geo_location['found'] != 0:
            geo[row] = [data_geo_location['results'][0]['LATITUDE'],
                        data_geo_location['results'][0]['LONGITUDE']
                       ]
        else:
            missing.append(row)

    def assign(postal):
        if postal in geo:
            x = geo[postal][0]
            return x
    i['Latitude'] = i['postal_code'].apply(assign)

    def assign(postal):
        if postal in geo:
            x = geo[postal][1]
            return x
    i['Longtitude'] = i['postal_code'].apply(assign)
```

```python
# get the nearest schools
from geopy.distance import geodesic

def find_nearest(add_lat_lon_start,add_lat_lon_end):
    
    results = {}
    # first column must be address
    for index,flat in enumerate(add_lat_lon_start.iloc[:,13]):
        
        # 2nd column must be latitude, 3rd column must be longitude
        flat_loc = (add_lat_lon_start.iloc[index,10],add_lat_lon_start.iloc[index,11])
        nearest_stn = ['',10]
        for ind, stn in enumerate(add_lat_lon_end.iloc[:,0]):
            stn_loc = (add_lat_lon_end.iloc[ind,6],add_lat_lon_end.iloc[ind,7])
            distance = geodesic(flat_loc,stn_loc).km

            if distance < nearest_stn[1]:
                nearest_stn[0] = stn
                nearest_stn[1] = distance

        results[flat] = nearest_stn
        
    return results
```

```python
#primary school
b = find_nearest(df,primary)

def assign(dist):
    if dist in b:
        x = b[dist][0]
        return x
df['nearest_primary_school'] = df['full_address'].apply(assign)

def assign(dist):
    if dist in b:
        x= b[dist][1]
        return x
df['nearest_primary_school(km)'] = df['full_address'].apply(assign)
```

```python
#secondary school
c = find_nearest(df,secondary)

def assign(dist):
    if dist in c:
        x = c[dist][0]
        return x
df['nearest_secondary_school'] = df['full_address'].apply(assign)

def assign(dist):
    if dist in c:
        x= c[dist][1]
        return x
df['nearest_secondary_school(km)'] = df['full_address'].apply(assign)
```

```python
#mixed level
d = find_nearest(df,mixed)

def assign(dist):
    if dist in d:
        x = d[dist][0]
        return x
df['nearest_mixed_level'] = df['full_address'].apply(assign)

def assign(dist):
    if dist in d:
        x= d[dist][1]
        return x
df['nearest_mixed_level(km)'] = df['full_address'].apply(assign)
```

```python
#JC_MI
e = find_nearest(df,JC_MI)

def assign(dist):
    if dist in e:
        x = e[dist][0]
        return x
df['nearest_JC/MI'] = df['full_address'].apply(assign)

def assign(dist):
    if dist in e:
        x= e[dist][1]
        return x
df['nearest_JC/MI(km)'] = df['full_address'].apply(assign)
```
