## KNN prediction for each town

## why the heck are accuracy, f1 and precision all the same??
https://stackoverflow.com/questions/54068401/can-the-precision-recall-and-f1-be-the-same-value

```python
a = {'accuracy':[],'f1':[],'precision':[]}

for i in l:
    df2 = df[df['town'] == i]
    x = df2[['sqrt_resale_price','sqrt_floor_area_sqm']]
    y = df2['flat_type']
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,
    random_state = 1, stratify = y)
    knn = KNeighborsClassifier(n_neighbors = 3)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    a['accuracy'].append(accuracy_score(y_test,y_pred))
    a['f1'].append(f1_score(y_pred, y_test, average ='micro'))
    a['precision'].append(precision_score(y_pred, y_test, average ='micro'))
    
df3 = pd.DataFrame(data = a ,columns = ['accuracy','f1','precision'],index =l) 
df3
```

## prediction of flat_type per town, applying bagging

```python
b = {'accuracy':[],'f1':[],'precision':[]}
for i in l:
    df2 = df[df['town'] == i]
    x = df2[['sqrt_resale_price','sqrt_floor_area_sqm']]
    y = df2['flat_type']
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,
    random_state = 1, stratify = y)
    #Create the kNN base classifier
    knn = KNeighborsClassifier(n_neighbors=3)

    #Create the Bagging classifier. Default base classifiers is Decision Tree. 
    # - n_estimator is the number of base classifiers (i.e. weak learners)
    model = BaggingClassifier(n_estimators=50, base_estimator=knn)

    #Fit the training feature Xs and training label Ys
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    b['accuracy'].append(accuracy_score(y_test,y_pred))
    b['f1'].append(f1_score(y_pred, y_test, average ='micro'))
    b['precision'].append(precision_score(y_pred, y_test, average ='micro'))

df4 = pd.DataFrame(data = b ,columns = ['accuracy','f1','precision'],index =l) 
df4
```

```python
c = {'accuracy':[],'f1':[],'precision':[]}
for i in l:
    df2 = df[df['town'] == i]
    x = df2[['sqrt_resale_price','sqrt_floor_area_sqm']]
    y = df2['flat_type']
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1, stratify = y)
    #Create the AdaBoost classifier. Default base classifiers is Decision Tree. 
    # - n_estimator is the number of base classifiers (i.e. weak learners)
    # - learning_rate controls the weight adjustments of each base classifiers. Default is 1
    model = AdaBoostClassifier(n_estimators=50,learning_rate=0.1)

    #Fit the training feature Xs and training label Ys
    model.fit(x_train, y_train)
    #Use the trained model to predict the test data
    y_pred = model.predict(x_test)
    c['accuracy'].append(accuracy_score(y_test,y_pred))
    c['f1'].append(f1_score(y_pred, y_test, average ='micro'))
    c['precision'].append(precision_score(y_pred, y_test, average ='micro'))
    #c[i]={'accuracy':accuracy_score(y_test,y_pred),'f1':f1_score(y_pred, y_test, average ='micro'),'precision':precision_score(y_pred, y_test, average ='micro')}

#c
```

## Find outliers based on (s.f.a.s.) and (s.r.p.) for each town and flat_type
fewer outliers allow for better prediction, since less spread

```python
df_all2 = copy.deepcopy(df_all)
```

```python
def three_sd_range(series): # series refers to a one column dataframes
    sd = series.std()# find the standard deviation
    mean = series.mean()# find the mean
    low = mean - 3 * sd # what is the lower bound of values to accept?
    high = mean + 3 * sd # what is the higher bound of values to accept?
    return (low, high)
```

```python
# steps: 
# 1. iterate through datatable in df_all2
for each_df in df_all2:
    #iterate through the two columns
    for col_name in ['sqrt_resale_price','sqrt_floor_area_sqm']:
# 2. retrive the lower and upper bounds of the data
        lower, upper = three_sd_range(each_df[col_name])

    # 3. find out if outliers exist and obtain their values
        has_outlier = (each_df[col_name].min() < lower) or (each_df[col_name].max() > upper)
        print('[Before] '+col_name + ' has outliers:' +str(has_outlier))

    # 4. remove the outliers
        if has_outlier:
            each_df.drop(each_df[each_df[col_name] < lower].index, inplace = True)
            each_df.drop(each_df[each_df[col_name] > upper].index, inplace = True)
```

## Normalisation

How to improve KNN?
For better results, normalizing data on the same scale is highly recommended. Generally, the normalization range considered between 0 and 1. KNN is not suitable for the large dimensional data. In such cases, dimension needs to reduce to improve the performance. Also, handling missing values will help us in improving results.

It is a common practice to normalize your data before feeding it to any machine learning algorithm.
While applying StandardScaler, each feature of your data should be normally distributed such that it will scale the distribution to a mean of zero and a standard deviation of one.

## PCA (Only for numerical columns)

Principle Component Analysis (PCA) is a orthogonal transformation to transform your high dimensional data into linear combination of your features. It is a form of feature extraction - To find new features to transform the data from a high dimensional space to a low dimensional space.

Check this link out for more examples: https://www.datacamp.com/community/tutorials/principal-component-analysis-in-python

In this section, we will be exploring PCA from the scikit-learn library.

```python
from sklearn.decomposition import PCA
for each_df in df_all2:
    # initialise PCA with _ principle components
    pca = PCA(n_components = 5)
    # drop all rows with null values, PCA cannot be done on null values
    each_df = each_df.dropna()
    each_df = each_df.reset_index(drop=True)
    # apply PCA for _____ columns 
    pca_components = pca.fit_transform(each_df.loc[:,['sqrt_resale_price','sqrt_floor_area_sqm','lease_commence_date','Year','Month']])
    # convert numpy array to dataframe
    pca_df = pd.DataFrame(pca_components, columns=["PCA(srp)", "PCA(sfas)","PCA(lcd)","PCA(Year)","PCA(Month)"])
    # combine the principle components into the original dataframe
    main_df = pd.concat([each_df, pca_df], axis=1)
    print('Variation by principal component: {}'.format(pca.explained_variance_ratio_))
#seems like the three major components are "PCA(srp)", "PCA(sfas)","PCA(lcd)"
```

## Below holds the best K for each year from 1990 to 1999

```python
#K_error_dict is a dictionary that holds all the error rate of K from 1 to 10 in a list from 1990 to 1999
from sklearn.model_selection import train_test_split
K_error_dict = {}
start = 1990
for each_df in df_all2:
    df = each_df
    #x = df[['sqrt_resale_price','sqrt_floor_area_sqm','lease_commence_date']]
    x = df[['Month','town_encoded','block_encoded','street_name_encoded','storey_range_encoded','sqrt_floor_area_sqm','flat_model_encoded','lease_commence_date','sqrt_resale_price']]
    y = df['flat_type']
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,
    random_state = 1, stratify = y)

    #calculating error for K from 1 to 10
    error = []
    for i in range(1, 10):
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(x_train, y_train)
        pred_i = knn.predict(x_test)
        error.append(np.mean(pred_i != y_test))
    K_error_dict[start] = error
    
    start = start + 1
```

- [Bagging Classifier SKLearn Documentation](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.BaggingClassifier.html)

*Note that the default AdaBoost implementation in SKLearn is Decision Tree 

```python
start = 1990
for i in range(len(df_all2)):
    df = df_all2[i]
    #x = df[['sqrt_resale_price','sqrt_floor_area_sqm','lease_commence_date']]
    x = df[['Month','town_encoded','block_encoded','street_name_encoded','storey_range_encoded','sqrt_floor_area_sqm','flat_model_encoded','lease_commence_date','sqrt_resale_price']]
    y = df['flat_type']
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,
    random_state = 1)    # - n_estimator is the number of base classifiers (i.e. weak learners)
    model = BaggingClassifier(n_estimators=50)

    #Fit the training feature Xs and training label Ys
    model.fit(x_train, y_train)
    #Use the trained model to predict the test data
    y_pred = model.predict(x_test)
    
    print('Year:'+str(start))
    start = start + 1
    # Find the accuracy and F1 score of the result 
    asr = accuracy_score(y_pred, y_test)
    f1 = f1_score(y_pred, y_test, average='micro')
    print(confusion_matrix(y_pred, y_test))
    print('accuracy:',asr)
    print('f1_score:',f1)
    print('\n')
```

## AdaBoost (with Decision Tree)

Refer to the following links on for detail explanation on the implementation:
- [AdaBoost Classifier SKLearn Documentation](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostClassifier.html)
- [DataCamp Implementation](https://www.datacamp.com/community/tutorials/adaboost-classifier-python)
- [Setting Learning Rate and N Estimators](https://stats.stackexchange.com/questions/82323/shrinkage-parameter-in-adaboost)

*Note that the default AdaBoost implementation in SKLearn is Decision Tree 


```python
start = 1990
for i in range(len(df_all2)):
    df = df_all2[i]
    #x = df[['sqrt_resale_price','sqrt_floor_area_sqm','lease_commence_date']]
    x = df[['Month','town_encoded','block_encoded','street_name_encoded','storey_range_encoded','sqrt_floor_area_sqm','flat_model_encoded','lease_commence_date','sqrt_resale_price']]

    y = df['flat_type']
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,
    random_state = 1)    # - n_estimator is the number of base classifiers (i.e. weak learners)
    
    #Create the AdaBoost classifier. Default base classifiers is Decision Tree. 
    # - n_estimator is the number of base classifiers (i.e. weak learners)
    # - learning_rate controls the weight adjustments of each base classifiers. Default is 1
    model = AdaBoostClassifier(n_estimators=50,learning_rate=0.1)

    #Fit the training feature Xs and training label Ys
    model.fit(x_train, y_train)

    #Use the trained model to predict the test data
    y_pred = model.predict(x_test)
    
    print('Year:'+str(start))
    start = start + 1
    # Find the accuracy and F1 score of the result 
    asr = accuracy_score(y_pred, y_test)
    f1 = f1_score(y_pred, y_test, average='micro')
    print(confusion_matrix(y_pred, y_test))
    print('accuracy:',asr)
    print('f1_score:',f1)
    print('\n')
```
