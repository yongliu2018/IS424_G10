## Categorical Encoding

### label coder

```python
#LabelEncoder categorical ordinal data column "storey_range"
lbl = LabelEncoder()
lbl.fit(list(df2['storey_range'].values))
df2['storey_range'] = lbl.transform(list(df2['storey_range'].values))

#check Shape
print('shape df2: {}'.format(df2.shape))
```

```python
df2 = df2.drop(columns = ['Day of Week','Month','town','Year','resale_price'])
```

```python
df2.head()
```

### One Hot Encoder

```python
#OneHotEncoder categorical nominal data column "town","flat_type","street_name","flat_model","Day of Week"
cols = ["flat_type","street_name","flat_model"]
for i in cols:
    labelbinarizer = LabelBinarizer()
    x = labelbinarizer.fit_transform(df2[i])
    encoded_df = pd.DataFrame(x, columns=labelbinarizer.classes_)
    df2 = pd.concat([df2, encoded_df], axis=1)
```

```python
from sklearn.decomposition import PCA
# initialise PCA with _ principle components
pca = PCA(n_components = 3)
# drop all rows with null values, PCA cannot be done on null values
df2 = df2.dropna()
df2 = df2.reset_index(drop=True)
# apply PCA for _____ columns 
pca_components = pca.fit_transform(df2.loc[:,['floor_area_sqm','log1p_resale_price','years_left']])
# convert numpy array to dataframe
pca_df2 = pd.DataFrame(pca_components, columns=["PCA(s.f.a.s)","PCA(l.r.p)","PCA(y.l)"])
# combine the principle components into the original dataframe
main_df2 = pd.concat([df2, pca_df2], axis=1)
main_df2
print('Variation by principal component: {}'.format(pca.explained_variance_ratio_))
#well if we assume just components of numerical, then log1p_resale_price, and floor_area_sqm wins
```

```python
#Find outliers
def three_sd_range(series): # series refers to a one column dataframes
    # Your code here
    sd = series.std()# find the standard deviation
    mean = series.mean()# find the mean
    low = mean - 3 * sd # what is the lower bound of values to accept?
    high = mean + 3 * sd # what is the higher bound of values to accept?
    return (low, high)
```

```python
lower, upper = three_sd_range(df2['log1p_resale_price'])
lower, upper
```

```python
# Find non-outliers for log1p_resale_price
df2['log1p_resale_price'].loc[pd.DataFrame([df2['log1p_resale_price'] >= lower, df2['log1p_resale_price'] <= upper]).all()] \
.describe()
```

```python
x = df2[['log1p_resale_price','floor_area_sqm']]
y = df2['flat_type']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,
random_state = 1, stratify = y)
#y_train.value_counts()
```

```python
error = []

# Calculating error for K values between 1 and 10
for i in range(1, 10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred_i = knn.predict(x_test)
    error.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1, 10), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=5)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

```python
#best K is now 5
knn5 = KNeighborsClassifier(n_neighbors = 5)
knn5.fit(x_train, y_train)
y_pred = knn5.predict(x_test)
```

```python
#best K is now 9
knn9 = KNeighborsClassifier(n_neighbors = 9)
knn9.fit(x_train, y_train)
y_pred = knn9.predict(x_test)
```

##### What is the best k? Finding the optimal k is called tuning the hyperparameter. A handy tool is grid search. In scikit-learn, we use GridSearchCV, which trains out model multiple times on a range of values specificed with the param_grid parameter and computes cross validation score, so that we can check which of our values for the tested hyperparameter performed best.

```python
#finding the best k, considering I assumed it as 5
# tuning the hyperparameter, gridsearch is used 
from sklearn.model_selection import GridSearchCV
knnGS = KNeighborsClassifier()
param_grid = {'n_neighbors':np.arange(2,10)}
knn_gscv = GridSearchCV(knnGS, param_grid, cv=10)
knn_gscv.fit(x,y)
knn_gscv.best_params_
```
