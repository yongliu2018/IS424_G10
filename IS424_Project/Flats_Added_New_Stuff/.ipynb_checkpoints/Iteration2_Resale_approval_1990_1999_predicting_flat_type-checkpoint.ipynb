## Checking for Skewness

```python
df2.hist(figsize = (10, 10))
plt.tight_layout
```

1. If skewness is less than -1 or greater than 1, the distribution is skewed, but may or may not be considered high depending on the values.

2. If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.

3. If skewness is between -0.5 and 0.5, the distribution is approximately symmetrical.

```python
df.skew()
```

```python
# this is for resale_price
# Plot Histogram
sns.distplot(df2['resale_price'] , fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(df2['resale_price'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('resale_price distribution')

fig = plt.figure()
plt.figure(figsize=(5,5)) 

res = stats.probplot(df2['resale_price'], plot=plt)
plt.show()

print("Skewness: %f" % df2['resale_price'].skew())
print("Kurtosis: %f" % df2['resale_price'].kurt())
```

## Normalisation of numerical data columns

It is a common practice to normalize your data before feeding it to any machine learning algorithm.

While applying StandardScaler, each feature of your data should be normally distributed such that it will scale the distribution to a mean of zero and a standard deviation of one.

```python
standardScaler = StandardScaler()

df3.loc[:,['sqrt_resale_price','sqrt_floor_area_sqm','sqrt_used_years']] = standardScaler.fit_transform(df3.loc[:,['sqrt_resale_price','sqrt_floor_area_sqm','sqrt_used_years']])
df3
```

## Factor analysis of mixed data (FAMD)
https://github.com/MaxHalford/prince#factor-analysis-of-mixed-data-famd

The available parameters are:

1. n_components: the number of components that are computed. You only need two if your intention is to make a chart.
2. n_iter: the number of iterations used for computing the SVD
3. rescale_with_mean: whether to substract each column's mean
4. rescale_with_std: whether to divide each column by it's standard deviation
5. copy: if False then the computations will be done inplace which can have possible side-effects on the input data
6. engine: what SVD engine to use (should be one of ['auto', 'fbpca', 'sklearn'])
7. random_state: controls the randomness of the SVD results.

```python
## Import libraries
import pandas as pd
import prince
import pprint

## Instantiate FAMD object
famd = prince.FAMD(
     n_components=9,
     n_iter=10,
     copy=True,
     check_input=True,
     engine='auto',       ## Can be "auto", 'sklearn', 'fbpca'
     random_state=42)

## Fit FAMD object to data 
famd = famd.fit(df3.drop('flat_type', axis=1)) ## Exclude target variable "flat_type"

## Inspect principal dimensions
pp = pprint.PrettyPrinter()
pp.pprint(famd.explained_inertia_) 
```

## PCA

```python
from sklearn.decomposition import PCA
# initialise PCA with _ principle components
pca = PCA(n_components = 3)
# apply PCA for _____ columns 
pca_components = pca.fit_transform(df3.loc[:,['sqrt_resale_price','sqrt_floor_area_sqm','sqrt_used_years']])
# convert numpy array to dataframe
pca_df = pd.DataFrame(pca_components, columns=["PCA(srp)", "PCA(sfas)","PCA(suy)"])
# combine the principle components into the original dataframe
main_df = pd.concat([df3, pca_df], axis=1)
main_df
print('Variation by principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
#Find outliers
def three_sd_range(series): # series refers to a one column dataframes
    # Your code here
    sd = series.std()# find the standard deviation
    mean = series.mean()# find the mean
    low = mean - 3 * sd # what is the lower bound of values to accept?
    high = mean + 3 * sd # what is the higher bound of values to accept?
    return (low, high)
```
