```python
from sklearn.decomposition import PCA
# initialise PCA with _ principle components
pca = PCA(n_components = 5)
# drop all rows with null values, PCA cannot be done on null values
df = df.dropna()
df = df.reset_index(drop=True)
# apply PCA for _____ columns 
pca_components = pca.fit_transform(df.loc[:,['sqrt_resale_price','sqrt_floor_area_sqm','lease_commence_date','Year','Month']])
# convert numpy array to dataframe
pca_df = pd.DataFrame(pca_components, columns=["PCA(srp)", "PCA(sfas)","PCA(lcd)","PCA(Year)","PCA(Month)"])
# combine the principle components into the original dataframe
main_df = pd.concat([df, pca_df], axis=1)
main_df
print('Variation by principal component: {}'.format(pca.explained_variance_ratio_))
#9.97707978e-01 -> 9.9771(10**-1) -> 0.99771 (sqrt_resale_price)
#1.40015208e-03 -> 1.4002(10**-3) -> 0.0014002 (sqrt_floor_area_sqm)
#6.09639273e-04 -> 6.0964(10**-4) -> 0.00060964 (lease_commence_date)
#2.61356212e-04 -> 2.6136(10**-4) -> 0.00026136 (Year)
#2.08742850e-05 -> 2.0874(10**-5) -> 0.000020874 (Month)
#So with pca we can tell that mainly resale_price and sqrt_floor_area_sqm will be the principal component
```

```python
#Find outliers
def three_sd_range(series): # series refers to a one column dataframes
    # Your code here
    sd = series.std()# find the standard deviation
    mean = series.mean()# find the mean
    low = mean - 3 * sd # what is the lower bound of values to accept?
    high = mean + 3 * sd # what is the higher bound of values to accept?
    return (low, high)
```
